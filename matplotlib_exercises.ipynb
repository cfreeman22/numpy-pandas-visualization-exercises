```python
#import pyplot module as alias plt
import matplotlib.pyplot as plt  

#import other libraries 
import numpy as np
import math
from random import randint
import random
import pandas as pd
```

### - 3 Combine the figures you created in the last step into one large figure with 4 subplots.

```python
# plot different mathematical functions vs x in a single figure
x = [x  for x in range(0,26)]
#pi = math.pi

y1 = [(math.sqrt(x)) for x in x]
y2 = [(x**3) for x in x]
y3 = [(2**x) for x in x]
y4 = [(1/(x+1)) * -1 for x in x]

plt.figure(figsize = (8,8))

plt.suptitle('Various Math Functions', fontsize = 16)

plt.subplot(2,2,1)
plt.plot(x,y1, color = 'r', marker = '*')
plt.title('$y =âˆšx $')
#plt.xticks([-2 * pi, -1 * pi, 0, pi, 2*pi], ['$2\pi$', '$\pi$', '0', '$\pi$', '$2\pi$'])

plt.subplot(2,2,2)
plt.plot(x, y2, color= 'b', marker = 's')
plt.title('$y =x3 $')



plt.subplot(2,2,3)
plt.plot(x, y3, color = 'g', marker = 'v')
plt.title('$y =2^x $')




plt.subplot(2,2,4)
plt.plot(x, y4, color = 'm', marker = 'o')
plt.title('$y =1/(x+1)$ ')
plt.axhline(0, color = 'm', lw = 0.5)
plt.fill_between(x, y4, color = 'm')

plt.tight_layout()
```
