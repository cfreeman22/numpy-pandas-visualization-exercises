```python
#import pyplot module as alias plt
import matplotlib.pyplot as plt  

#import other libraries 
import numpy as np
import math
from random import randint
import random
import pandas as pd
```

- Title your chart "Big O Notation"
- Label your x axis "Elements"
- Label your y axis "Operations"
- Label your curves or make a legend for the curves
- Use LaTex notation where possible

```python
x = range(1,5)

y1 = [0*n + 1 for n in x]
plt.figure(figsize = (10,8))

#y1 = [(math.sqrt(i)) for i in x]
y2 = [math.log2(a) for a in x]
y3 = [b for b in x]
y4 = [c*(math.log2(c)) for c in x]
y5 = [(d**2) for d in x]
y6 = [(2**e) for e in x]
y7 = [(math.factorial(f)) for f in x]
y8 = [(g**g) for g in x]

plt.xlabel('Elements')
plt.ylabel('Operations')
plt.plot(x,y1, label='O(n) + 1')
plt.plot(x, y2,label='(log n)' )
plt.plot(x,y3, label='O(n)')
plt.plot(x, y4, label='(n log n)') 
plt.plot(x,y5, label='(n^2)')
plt.plot(x, y6,label='(2^n)' )
plt.plot(x,y7, label='(n!)')
#plt.plot(x, y8, label='y = O(n^n)')
plt.title('Big O Notation')
plt.legend()
```
